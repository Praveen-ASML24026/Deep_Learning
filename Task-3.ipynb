```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization, LayerNormalization
```

```python
np.random.seed(42)

data = np.random.randint(1, 20, size=(4, 5)).astype(np.float32)

print("Original Input Values:")
print(data)
```

```python
batch_norm = BatchNormalization()

batch_output = batch_norm(data, training=True)

print("Batch Normalization Output:")
print(batch_output.numpy())
```

```python
training=True
```

```python
layer_norm = LayerNormalization()

layer_output = layer_norm(data)

print("Layer Normalization Output:")
print(layer_output.numpy())
```
